```python
import requests
from bs4 import BeautifulSoup
import time
person_img_set_lsit = []
list_url =[]
list_name =[]
```

```python
url = "https://www.meitulu.com/search/闫盼盼"
```

```python
response = requests.get(url)
print(response.text)
```

```python
soup = BeautifulSoup(response.text,"html.parser")
print(soup)
```

```python
person_img_sets = soup.select(".boxs .img li .p_title a")
print(person_img_sets)
```

```python
for i,item in enumerate(person_img_sets):
    person_img_set_url = item.get("href")
    person_img_set_name = str(i+1)
    print(person_img_set_url,",",person_img_set_name)
    dic  ={"person_img_set":{"person_img_set_name":person_img_set_name,"person_img_set_url":person_img_set_url}}
    person_img_set_lsit.append(dic)
    list_url.append(person_img_set_url)
    list_name.append(person_img_set_name)
    
#     print(dic)
# print(list_name)
```
